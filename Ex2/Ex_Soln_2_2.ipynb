```python
import numpy as np
from scipy.misc import imread, imsave, imresize
import skimage
from skimage import data
from skimage.feature import corner_harris
from skimage.feature import peak_local_max
import matplotlib.pyplot as plt
from skimage import data,exposure, img_as_float
import matplotlib.pyplot as plt
from skimage import io
%matplotlib inline
```

```python
def histogram_equalize(img):
    cdf, bins = exposure.cumulative_distribution(img,nbins=10)
    
    print 'cdf : ',cdf
    
    print 'nbins :',bins
    plt.plot(bins,cdf)
    plt.show()
    
    
    
    
    
    return np.interp(img, bins, cdf )
```

```python
img= img_as_float(io.imread('myelin.png'))

hist_img_org , nbins_img_org = skimage.exposure.histogram(img,nbins=10)
plt.plot(nbins_img_org, hist_img_org)
```

```python
img_he= histogram_equalize(img)
hist_img, bins_img= exposure.histogram(img, nbins=10)
hist_img_he, bins_img_he=exposure.histogram(img_he, nbins=20)




plt.figure(1, figsize=(8,8))
plt.subplot(221)
plt.imshow(img, cmap='gray')
plt.subplot(222)



plt.imshow(img_he, cmap='gray')
plt.subplot(223)
plt.plot(bins_img, hist_img)
plt.subplot(224)
plt.plot(bins_img_he, hist_img_he)
#plt.show()
```
